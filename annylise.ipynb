# Q1
1. Compare the average delivery time for each product category, and highlight which category has the fastest and slowest average delivery.

# â†’what i need to do first
1. i have to pick delvery time columns and have to pick product category column and need to find mean
2. if i have to higlight fastest category and slowes average delivery so i have to first pick top 5 product that delve fast
3. and have to pick lower product whic have delvery delay 


```python
baba = df.groupby('Product Category')['Service Rating'].mean()
colors = sns.color_palette('viridis',len(baba))
for spine in plt.gca().spines.values():
    spine.set_linewidth(3)


def label(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],f"{y[i]:.1f}",ha='center',va='bottom')
plt.bar(baba.index,baba.values,color=colors,edgecolor='black',linewidth=1.2)
label(baba.index,baba.values)
plt.xticks(rotation=40)
plt.grid(True,linewidth=2)
plt.title('Bar Plot Of Product & Category')
plt.ylabel('Product')
plt.tight_layout()
plt.show()
```

```python
delevery1 = df.groupby('Product Category')['Delivery Time (Minutes)'].mean()
delevery2= delevery1.sort_values()
for spine in plt.gca().spines.values():
    spine.set_linewidth(2)
def delevery(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],f"{y[i]:.1f}",ha='center',va='bottom')
colors=sns.color_palette('viridis',len(delevery2))
colors = list(colors)
colors[0] = (0.2, 0.8, 0.2)  # Green for fastest
colors[-1] = (0.9, 0.2, 0.2)
plt.bar(delevery2.index,delevery2.values,color=colors,edgecolor='black')
delevery(delevery2.index,delevery2.values)
plt.title('average delivery time for each product category',fontsize=17)
plt.ylabel('Product')
plt.tight_layout()
plt.grid(True,linewidth=2)
plt.xticks(rotation=40)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
comparision = df.groupby('Platform')['Service Rating'].mean().sort_values(ascending=False)
colors = sns.color_palette('viridis', len(comparision))
colors = list(colors)
colors[0] = (0.2, 0.8, 0.2)  
colors[-1] = (0.9, 0.2, 0.2) 
explode = [0.05 if i in [0, len(comparision)-1] else 0 for i in range(len(comparision))]

fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    comparision.values,
    labels=comparision.index,
    autopct=lambda pct: f"{pct:.1f}%\n({int(pct/100*comparision.sum())})",
    colors=colors,
    explode=explode,
    shadow=True,
    startangle=140,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1},
    textprops={'fontsize': 12}
)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

plt.title('Average Service Rating by Platform', fontsize=16, fontweight='bold', pad=20)
plt.legend(wedges, comparision.index, title="Platforms", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Numeric conversion
delivery_time = pd.to_numeric(df['Delivery Time (Minutes)'], errors='coerce')
order_value = pd.to_numeric(df['Order Value (INR)'], errors='coerce')

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Scatter plot with style
scatter = plt.scatter(
    x=delivery_time,
    y=order_value,
    c=order_value,                  # Color by order value
    s=delivery_time * 5,            # Bubble size by delivery time
    cmap='viridis',
    alpha=0.7,
    edgecolors='black',
    linewidths=0.5
)

# Titles & labels
plt.title("Delivery Time vs Order Value", fontsize=16, fontweight='bold')
plt.xlabel("Delivery Time (Minutes)", fontsize=12)
plt.ylabel("Order Value (INR)", fontsize=12)

# Colorbar
cbar = plt.colorbar(scatter)
cbar.set_label("Order Value (INR)", fontsize=12)

# Annotate highest and lowest orders
max_idx = order_value.idxmax()
min_idx = order_value.idxmin()
plt.annotate(f"Max Order\nâ‚¹{order_value[max_idx]}",
             (delivery_time[max_idx], order_value[max_idx]),
             xytext=(delivery_time[max_idx]+5, order_value[max_idx]+200),
             arrowprops=dict(facecolor='black', arrowstyle="->"),
             fontsize=10, fontweight='bold')

plt.annotate(f"Min Order\nâ‚¹{order_value[min_idx]}",
             (delivery_time[min_idx], order_value[min_idx]),
             xytext=(delivery_time[min_idx]+5, order_value[min_idx]+200),
             arrowprops=dict(facecolor='black', arrowstyle="->"),
             fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```
